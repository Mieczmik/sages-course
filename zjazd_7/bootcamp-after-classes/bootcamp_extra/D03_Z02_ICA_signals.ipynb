````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal

from sklearn.decomposition import PCA, FastICA
````

# ICA

## Zad.

Wygeneruj dwa sygna≈Çy.

````python
# #############################################################################
# Generate sample data
np.random.seed(0)
n_samples = 2000
time = np.linspace(0, 8, n_samples)

s1 = np.sin(20 * time)  # Signal 1 : sinusoidal signal
s2 = signal.sawtooth(2 * np.pi * time)  # Signal 2 : square signal

S = np.c_[s1, s2]
````

````python
S /= S.std(axis=0)  # Standardize data

plt.figure()
df = pd.DataFrame(S)
df.plot(subplots=True, figsize=(10, 5));
plt.show()
````

````python
plt.figure(figsize=(8,8))
plt.scatter(S[:, 0], S[:, 1])
plt.xlim([-2,2])
plt.ylim([-2,2])
plt.axis('equal')
plt.show()
````

# Zad 

dokonaj mieszania sygna≈Ç√≥w.

````python
# Mix data
A = np.array([[0.2, -0.1], [0.5, 2]])  # Mixing matrix
A
````

# Zad
Narysuj sygna≈Çy po odmieszaniu.

````python
X = np.dot(S, A.T)  # Generate observations

plt.figure()
df = pd.DataFrame(X)
df.plot(subplots=True, figsize=(10, 5));
plt.show()
````

````python
plt.figure(figsize=(8,8))
plt.scatter(X[:, 0], X[:, 1])
plt.xlim([-4,4])
plt.ylim([-4,4])
plt.axis('equal')
plt.show()
````

# Zad
Wykonaj algorytm PCA i wizualizuj bazƒô (komponenty)


````python
pca = PCA()
S_pca_ = pca.fit(X).transform(X)
````

````python
plt.figure(figsize=(8,8))
plt.scatter(X[:, 0], X[:, 1])
x_axis, y_axis = pca.components_.T
plt.quiver(0, 0, x_axis, y_axis, zorder=11, width=0.01, scale=6,
                       color="r")

plt.xlim([-4,4])
plt.ylim([-4,4])
plt.axis('equal')
plt.show()
````

# Zad.
Wykonaj analogiczne zadanie na sygna≈Çach z szumem
 

````python
# #############################################################################
# Generate sample data
np.random.seed(0)
n_samples = 2000
time = np.linspace(0, 8, n_samples)

s1 = np.sin(2 * time)  # Signal 1 : sinusoidal signal
s2 = np.sign(np.sin(3 * time))  # Signal 2 : square signal


S = np.c_[s1, s2]
S += 0.2 * np.random.normal(size=S.shape)  # Add noise
````

````python
S /= S.std(axis=0)  # Standardize data

plt.figure()
df = pd.DataFrame(S)
df.plot(subplots=True, figsize=(10, 5));
plt.show()
````
