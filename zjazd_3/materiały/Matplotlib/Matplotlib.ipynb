## Zaawansowane stylowanie wykresu

### Zaawansowane stylowanie ticków

```python
data = np.random.rand(20)
```

```python
fig, ax = plt.subplots(figsize = (10,7))

ax.set_title("Tytuł", {'fontsize':20})
ax.plot(data, 'b--', linewidth = 2)
ax.set_xlabel('Oś x', fontsize=16)
ax.set_ylabel('Oś y', fontsize=16)

ax.tick_params(axis='both', which='both', direction='in', labelsize=14)  # axis: x, y, both  which: major, minor, both
ax.minorticks_on()
```

### Dwie osie pionowe

```python
t = np.arange(0.01, 10.0, 0.01)
data1 = np.exp(t)
data2 = np.sin(2 * np.pi * t)
```

```python
fig, ax1 = plt.subplots()

ax1.set_xlabel('time (s)', fontsize=14)
ax1.set_ylabel('exp', color='red', fontsize=14)
ax1.plot(t, data1, color='red')
ax1.tick_params(axis='y', labelcolor='red')

ax2 = ax1.twinx()

ax2.set_ylabel('sin', color='blue', fontsize=14)
ax2.plot(t, data2, color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

ax1.set_xticks([2, 4, 6, 8])
ax1.set_xticklabels(['2', '4', '6', '8'], fontdict={"fontsize": 18})

plt.show()
```

### Współdzielenie osi

```python
fig, axes = plt.subplots(figsize=(10, 7), sharex=True, sharey=True, nrows=2, ncols=2)

axes[0, 0].plot([1, 2, 3])
axes[0, 1].plot([4, 4, 2])
axes[1, 0].plot([8, 4, 1])
axes[1, 1].plot([3, 6, 1])

axes[0][0].set_title("Tytuł wykresu 0, 0", fontsize=15)
axes[0][1].set_title("Tytuł wykresu 0, 1", fontsize=15)
axes[1][0].set_title("Tytuł wykresu 1, 0", fontsize=15)
axes[1][1].set_title("Tytuł wykresu 1, 1", fontsize=15)

plt.suptitle("Tytuł całego wykresu", fontsize=18)
    
plt.subplots_adjust(left=0.1, bottom=0.05, right=0.9, top=0.9, wspace=0.15, hspace=0.35)
```

###### Zadanie 1
Odtwórz wykres korzystając z utworzonych poniżej wektorów



```python
x = np.linspace(0, 10, 1000)
y_1 = np.log(x+1) + np.sin(x/2)**2
y_2 = np.sin(2*x**2) - np.sqrt(x*np.log(2*x+1))
y_3 = 10*np.sin(x)
```

```python
fig, ax1 = plt.subplots(figsize=(13, 3))

ax1.plot(x, y_1)
ax1.plot(x, y_2)

ax2 = ax1.twinx()

ax2.plot(x[::10], y_3[::10], 'r.')

ax1.set_yticks([-4, -6, 0, 2, 3])
ax1.tick_params('y', rotation = -30)
ax1.tick_params('x', rotation = 30)

ax2.set_yticks([-8, -5, 0, 5, 8])
ax2.grid()

ax1.set_ylabel('Oś Y1', fontsize=14)
ax2.set_ylabel('Oś Y2', fontsize=14)
ax1.set_xlabel('Oś X', fontsize=14)

fig.suptitle("Tytuł wykresu", fontsize=18)
```

```python
fig, ax1 = plt.subplots(figsize=(30, 8))

ax1.plot(x, y_1)
ax1.plot(x, y_2)

ax2 = ax1.twinx()
ax2.plot(x[::10], y_3[::10], 'o', c='r')

ax1.set_xlabel("Oś X", fontsize=20)
xticks = [2, 4, 6, 8]
ax1.set_xticks(xticks)
ax1.set_xticklabels([str(i) for i in xticks], fontdict={'fontsize': 24, 'rotation': 30})
ax1.tick_params(axis='x', which='both', direction='in', size=20)

y1_ticks = [-6, -4, 0, 2, 3]
ax1.set_yticks(y1_ticks)
ax1.set_yticklabels([str(i) for i in y1_ticks], fontdict={'fontsize': 20, 'rotation': -30})
ax1.set_ylabel("Oś Y1", fontsize=20)

y2_ticks = [-8, -5, 0, 5, 8]
ax2.set_yticks(y2_ticks)
ax2.set_yticklabels([str(i) for i in y2_ticks], fontdict={'fontsize': 20})
ax2.set_ylabel("\n\nOś Y2", fontsize=20)

plt.title("Tytuł wykresu", fontsize=25)

ax2.grid()

plt.show()
```
