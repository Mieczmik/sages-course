# Rozkład wykładniczy - $\text{Exp}(\lambda)$
Zmienna losowa $X$ ma rozkład wykładniczy z parametrem $\lambda>0$, jeżeli jej gęstość $f$~jest postaci
$$
      f(x)=\left.
    \begin{cases}
    \lambda\exp\{-\lambda x\} \quad &\text{dla} \quad x>0\\
    0 \quad &\text{dla} \quad x\leq 0,
    \end{cases}\right.
$$

Wartość oczekiwana i wariancja dane są wzorami
$$
    \mathbb{E}X = \frac{1}{\lambda}, \quad \text{Var}(X) = \frac{1}{\lambda^2}.
$$

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html#scipy.stats.expon

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy import stats
```

6. Narysujmy histogram dla wylosowanej powyżej próbki.

```python
l = 2
exp2 = stats.expon(scale = 1 / l)# scale = wartosc oczekiwana = 1 / lambda
exp2
```

```python
# Gęstość

x = np.linspace(-1, 10, 1000)
plt.plot(x, exp2.pdf(x), label = "exp(2) pdf")
plt.plot(x, stats.expon(scale = 0.25).pdf(x), label = "exp(4) pdf") # lambda = 4
plt.plot(x, stats.expon(scale = 4).pdf(x), label = "exp(0.25) pdf") # lambda = 0.25
plt.legend()
plt.show()
```

```python
# Dystrubuanta

x = np.linspace(-1, 10, 1000)
plt.plot(x, exp2.cdf(x), label = "exp(2) cdf")
plt.plot(x, stats.expon(scale = 0.25).cdf(x), label = "exp(4) cdf") # lambda = 4
plt.plot(x, stats.expon(scale = 4).cdf(x), label = "exp(0.25) cdf") # lambda = 0.25
plt.show()
```
