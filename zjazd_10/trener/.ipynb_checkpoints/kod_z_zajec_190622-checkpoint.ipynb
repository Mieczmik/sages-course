```python
import nltk
from bs4 import BeautifulSoup
import numpy as np

wiki = "http://en.wikipedia.org/wiki/"
titles = ["Integral", "Riemann_integral", "Riemann-Stieltjes_integral", "Derivative",
    "Limit_of_a_sequence", "Edvard_Munch", "Vincent_van_Gogh", "Jan_Matejko",
    "Lev_Tolstoj", "Franz_Kafka", "J._R._R._Tolkien"]

from urllib.request import urlopen

def parse(url):        
    x = urlopen(url).read()
    x = BeautifulSoup(x).find("div",id="bodyContent").find_all("p")
    return(x)

articles = [parse(url) for url in [wiki+x for x in titles]]
articles = [" ".join([str(paragraph) for paragraph in article]) for article in articles]

articles[0][:200]
```

```python
from sklearn.datasets import fetch_20newsgroups

categories = ['alt.atheism', 
              'talk.religion.misc',
              'comp.graphics', 
              'sci.space']
newsgroups_train = fetch_20newsgroups(subset='train', 
                                      categories=categories)
```

```python
texts = newsgroups_train.data
y = newsgroups_train.target
```

```python
texts_preprocessed = [nltk.word_tokenize(mail) for mail in texts]

stemmer = nltk.PorterStemmer()
texts_preprocessed = [[stemmer.stem(word) for word in mail] for mail in texts_preprocessed]

texts_preprocessed = [" ".join(mail) for mail in texts_preprocessed]
```

```python
print(texts_preprocessed[0])
```

```python
vectorizer = TfidfVectorizer()
vectorizer.fit(texts_preprocessed)
X = vectorizer.transform(texts_preprocessed)
words = np.array(vectorizer.get_feature_names())
```

```python
k = 7
i = 500
print(words[X[i].A.flatten().argsort()[::-1]][:k])
print("")
print(texts[i])
```
